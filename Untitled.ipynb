```python
import random
import numpy as np
```

```python
# Abstract struct class       
class Struct:
    def __init__ (self, *argv, **argd):
        if len(argd):
            # Update by dictionary
            self.__dict__.update(argd)
        else:
            # Update by position            
            for n in range(len(argv)):
                setattr(self, self.order[n], argv[n])

# Specific class
class Transaction(Struct):    
    t = None
    f = None    
    a = 0
    
    def __init__(self, *args, **kwargs):
        self.order = ['t', 'f', 'a']
          
        super(Transaction, self).__init__(*args, **kwargs)
        
    def __str__(self):
        return "t: %s f: %s a: %s" % (self.t, self.f, self.a)
```

```python
actors = [hex(x) for x in range(10**5)]
```

```python
def pick_actors(actors):
    t = random.choice(actors)
    f = random.choice(actors)
    
    while f == t:
        t = random.choice(actors)
        
    return t, f

def random_int():
    return (np.random.poisson(3) + 1) ** 4
```

```python
transactions = []

for _ in range(10**10):
    transactions.append(Transaction(*pick_actors(actors), random_int()))
```

```python
[x.__str__() for x in transactions]
```
